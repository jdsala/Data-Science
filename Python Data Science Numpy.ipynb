## **Loop Aninhado**

```python
dados = [ 
    ['A', 'B', 'C'],
    ['D', 'E', 'F'],
    ['G', 'H', 'I']
]
```

```python
result = []
for lista in dados:
    for item in lista:
        result.append(item)
result
```

```python
result_2 = []
for lista in dados:
    result_2 += lista
result_2
```

```python
[item for lista in dados for item in lista]
```
